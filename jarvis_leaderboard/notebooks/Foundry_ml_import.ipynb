```python
from foundry import Foundry
no_local_server = False
no_browser = False
globus = True
f = Foundry(no_local_server=no_local_server, no_browser=no_browser, index="mdf")
```

```python
f.load("10.18126/wg3u-g8vu", globus=False)
res = f.load_data()
#X_mp,y_mp = res['train'][0], res['train'][1]
```

```python
f
```

```python
X=res['train'][0]
X['id']=X.index
Y=res['train'][1]
Y['id']=Y.index
```

```python
import pandas as pd
df = pd.merge(X,Y,on='id')
```

```python
df2=df.sample(frac=1)
```

```python
mem=[]
for i,ii in df2.iterrows():
        try:
            id=str('fdml-expgap-')+str(i)
            info={}
            info['id']=id
            info['composition']=ii['composition']
            info['exp_gap']=float(ii['bandgap value (eV)'])
            mem.append(info)
        except:
            print(i,ii['bandgap value (eV)'])
            pass
```

```python
len(mem)
```

```python
test_ratio = 0.2
n_test=int(len(mem)*test_ratio)
n_train=len(mem)-n_test
```

```python
n_train,n_test
```

```python
from collections import defaultdict
train={}
test={}
for i,ii in enumerate(mem):
    if i<n_train:
        info={}
        train[ii['id']]=float(ii['exp_gap'])
    else:
        test[ii['id']]=float(ii['exp_gap'])
```

```python
from jarvis.db.jsonutils import dumpjson
m={}
m['train']=train
m['test']=test
#Add to https://github.com/usnistgov/jarvis_leaderboard/tree/main/jarvis_leaderboard/benchmarks/AI/SinglePropertyPrediction
dumpjson(data=m,filename='foundry_ml_exp_bandgaps_exp_gap.json')

#Add to https://github.com/usnistgov/jarvis/blob/master/jarvis/db/figshare.py#L26
dumpjson(data=mem,filename='foundry_ml_exp_bandgaps.json')
```
